```python
len(X.columns)
```

```python
model=Sequential()
model.add(Dense(11,activation="sigmoid",input_dim=11)) # input layer
model.add(Dense(11,activation="sigmoid"))  # hidder layer
model.add(Dense(1,activation="sigmoid"))  # outer layer
```

```python
#  11*11=121+11=132   1st hd
#  11*11=121+11=132   2nd hd
#  11*1+1 =12        3rd hd

#  132+132+12 =276   output
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler =  StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.fit_transform(X_test)
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train_scale,y_train,batch_size=50,epochs=10,verbose=1)    # testing accuracy
```

```python
# 1 epoch = 1FP + 1BP
```

```python
y_pred=model.predict(X_test_scale)
```

```python
y_pred
```

```python
y_pred=y_pred.argmax(axis=-1)
```

```python
from sklearn.metrics import accuracy_score       # treaing acccuracy
accuracy_score(y_test,y_pred)
```
